```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('graph.csv')
```

```python
df.head()
```

```python
from pandas import Series, DataFrame
s = df['Document Type'].str.split(';').apply(Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'Document Type'
del df['Document Type']
df = df.join(s)
s.head()
```

```python
df1 = df.groupby(['Document Type'],as_index=False)['Document Type'].agg({'Count':'count'})
df2 = df.groupby(['Document Type','Year Published'],as_index=False)['Document Type'].agg({'Count':'count'})
df1 = df1.sort_values('Count',ascending=False)
df2 = df2.sort_values('Year Published',ascending=False)
df1 = pd.DataFrame(data=df1)
df2 = pd.DataFrame(data=df2)
df1['Document Type'].replace('', np.nan, inplace=True)
df1.dropna(subset=['Document Type'], inplace=True)
df2['Document Type'].replace('', np.nan, inplace=True)
df2.dropna(subset=['Document Type'], inplace=True)
df1.index = np.arange(1,len(df1)+1)
df2.index = np.arange(1,len(df2)+1)
df1.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 8))
ax = sns.lineplot(x="Year Published",y="Count",hue="Document Type", data=df2)
```

```python
WOS_all = df['WOS Cited count']
print(WOS_all.describe())
f, ax= plt.subplots(figsize = (14, 10))
sns.distplot(WOS_all,ax = ax)
ax.set_title('WOS Cited Distribution',fontsize=25)
```

```python
uc180_all = df['UC 180']
print(uc180_all.describe())
f, ax= plt.subplots(figsize = (14, 10))
sns.distplot(uc180_all,ax = ax)
ax.set_title('Cited in 180 Days Distribution',fontsize=25)
```

```python
s_all = df['UC since 2013']
print(s_all.describe())
f, ax= plt.subplots(figsize = (14, 10))
sns.distplot(s_all,ax = ax)
ax.set_title('Cited since 2013 Distribution',fontsize=25)
```
