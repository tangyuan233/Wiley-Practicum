```python
# Load file
import pandas as pd
data = pd.read_csv('E:\OneDrive\Classes\BIA 686\data.csv',encoding="ISO-8859-1")
data.head(5)
```

```python
# Split Reserach Areas
from pandas import Series, DataFrame
s = data['Research Areas'].str.split(';').apply(Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'Research Areas'
del data['Research Areas']
data = data.join(s)
s.head()
```

```python
# Find Top 10 Authors
import numpy as np
df = pd.DataFrame(data=data)
df = df.groupby(['Authors'],as_index=False)['Authors'].agg({'Count':'count'})
df = df.sort_values('Count',ascending=False)
df= df.drop(df.index[0])
df.index = np.arange(1,len(df)+1)
df.head(10)
```

```python
# Find Top 10 Publisher
import numpy as np
df = pd.DataFrame(data=data)
df = df.groupby(['Publisher'],as_index=False)['Publisher'].agg({'Count':'count'})
df = df.sort_values('Count',ascending=False)
df= df.drop(df.index[0])
df.index = np.arange(1,len(df)+1)
df.head(10)
```

```python
# Find Top 10 Publisher City
import numpy as np
df = pd.DataFrame(data=data)
df = df.groupby(['Publisher City'],as_index=False)['Publisher City'].agg({'Count':'count'})
df = df.sort_values('Count',ascending=False)
df= df.drop(df.index[0])
df.index = np.arange(1,len(df)+1)
df.head(10)
```

```python
# Find Top 10 Author Country
import numpy as np
df = pd.DataFrame(data=data)
df = df.groupby(['Author Country'],as_index=False)['Author Country'].agg({'Count':'count'})
df = df.sort_values('Count',ascending=False)
df= df.drop(df.index[0])
df.index = np.arange(1,len(df)+1)
df.head(10)
```

```python
# Find Top 10 Research Areas
df1 = data.groupby(['Research Areas'],as_index=False)['Research Areas'].agg({'Count':'count'})
df2 = data.groupby(['Research Areas','Year Published'],as_index=False)['Research Areas'].agg({'Count':'count'})
df1 = df1.sort_values('Count',ascending=False)
df2 = df2.sort_values('Year Published',ascending=False)
df1 = pd.DataFrame(data=df1)
df2 = pd.DataFrame(data=df2)
df1['Research Areas'].replace('', np.nan, inplace=True)
df1.dropna(subset=['Research Areas'], inplace=True)
df2['Research Areas'].replace('', np.nan, inplace=True)
df2.dropna(subset=['Research Areas'], inplace=True)
df1.index = np.arange(1,len(df1)+1)
df2.index = np.arange(1,len(df2)+1)
df1.head(10)
```

```python
# Linelpot for all research areas over years
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 6))
ax = sns.lineplot(x="Year Published", y="Count", hue="Research Areas", data=df2)
```

```python
# Lineplot for Top 10 Research Areas over years
d1 = df2[df2['Research Areas'].str.contains("Materials Science")]
d2 = df2[df2['Research Areas'].str.startswith(" Engineering")]
d3 = df2[df2['Research Areas'].str.contains ("Chemistry")]
d4 = df2[df2['Research Areas'].str.contains ("Science & Technology - Other Topics")]
d5 = df2[df2['Research Areas'].str.contains ("Physics")]
d6 = df2[df2['Research Areas'].str.contains ("Polymer Science")]
d7 = df2[df2['Research Areas'].str.contains ("Biochemistry & Molecular Biology")]
d8 = df2[df2['Research Areas'].str.contains ("Pharmacology & Pharmacy")]
d9 = df2[df2['Research Areas'].str.contains ("Optics")]
d10 = df2[df2['Research Areas'].str.contains ("Computer Science")]
d_all = pd.concat([d1, d2, d3, d4, d5, d6, d7, d8, d9, d10])
d_all.head(5)
plt.figure(figsize=(16, 8))
ax = sns.lineplot(x="Year Published", y="Count", hue="Research Areas", data=d_all)
```
