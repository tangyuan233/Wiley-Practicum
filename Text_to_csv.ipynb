```python
#import records
import pandas as pd
import numpy as np
data = pd.read_csv("all.txt",delimiter="\t")
#1. PT - Publication Type
df1 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('PT')]
df1.index = np.arange(1,len(df1)+1)
df1.columns = ['Publication Type']
df1 = df1['Publication Type'].str.replace('PT', '')
df1 = pd.DataFrame(data=df1)

#2. AU - Authors
df2 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('AU')]
df2.index = np.arange(1,len(df2)+1)
df2.columns = ['Authors']
df2 = df2['Authors'].str.replace('AU', '')
df2 = pd.DataFrame(data=df2)

#3. TI - Document Title
df3 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('TI')]
df3.index = np.arange(1,len(df3)+1)
df3.columns = ['Title']
df3 = df3['Title'].str.replace('TI', '')
df3 = pd.DataFrame(data=df3)

#4. LA - Language
df4 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('LA')]
df4.index = np.arange(1,len(df4)+1)
df4.columns = ['Language']
df4 = df4['Language'].str.replace('LA', '')
df4 = pd.DataFrame(data=df4)

#5. DT - Document Type
df5 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('DT')]
df5.index = np.arange(1,len(df5)+1)
df5.columns = ['Document Type']
df5 = df5['Document Type'].str.replace('DT', '')
df5 = pd.DataFrame(data=df5)

#6. DE - Author Keywords
df6 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('DE')]
df6.index = np.arange(1,len(df6)+1)
df6.columns = ['Author Keywords']
df6 = df6['Author Keywords'].str.replace('DE', '')
df6 = pd.DataFrame(data=df6)

#7. RP - Reprint Address (only the country)
df7 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('RP')]
df7.index = np.arange(1,len(df7)+1)
df7.columns = ['Author Country']
df7['Author Country'] = df7.loc[:,'Author Country'].apply(lambda x: x[::-1])
df7['Author Country'] = df7['Author Country'].str.split().str.get(0)
df7['Author Country'] = df7.loc[:,'Author Country'].apply(lambda x: x[::-1])
df7['Author Country'] = df7['Author Country'].str.replace('.', '')
df7 = pd.DataFrame(data=df7)

#8. CR - Cited References
df8 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('CR')]
df8.index = np.arange(1,len(df8)+1)
df8.columns = ['Cited References']
df8 = df8['Cited References'].str.replace('CR', '')
df8 = pd.DataFrame(data=df8)

#9. NR - Cited Reference Count
df9 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('NR')]
df9.index = np.arange(1,len(df9)+1)
df9.columns = ['Cited Reference Count']
df9 = df9['Cited Reference Count'].str.replace('NR', '')
df9 = pd.DataFrame(data=df9)

#10. Z9 - Total Times Cited Count
df10 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('Z9')]
df10.index = np.arange(1,len(df10)+1)
df10.columns = ['CR count']
df10 = df10['CR count'].str.replace('Z9', '')
df10= pd.DataFrame(data=df10)

#11. TC - Web of Science Core Collection Times Cited Count
df11 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('TC')]
df11.index = np.arange(1,len(df11)+1)
df11.columns = ['WOS Cited count']
df11 = df11['WOS Cited count'].str.replace('TC', '')
df11 = pd.DataFrame(data=df11)

#12. WC - Web of Science Categories
df12 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('WC')]
df12.index = np.arange(1,len(df12)+1)
df12.columns = ['WOS Categories']
df12 = df12['WOS Categories'].str.replace('WC', '')
df12 = pd.DataFrame(data=df12)

#13. SC - Research Areas
df13 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('SC')]
df13.index = np.arange(1,len(df13)+1)
df13.columns = ['Research Areas']
df13 = df13['Research Areas'].str.replace('SC', '')
df13 = pd.DataFrame(data=df13)

#14. U1 - Usage Count (Last 180 Days)
df14 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('U1')]
df14.index = np.arange(1,len(df14)+1)
df14.columns = ['UC 180']
df14 = df14['UC 180'].str.replace('U1', '')
df14 = pd.DataFrame(data=df14)

#15. U2 - Usage Count (Since 2013)
df15 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('U2')]
df15.index = np.arange(1,len(df15)+1)
df15.columns = ['UC since 2013']
df15 = df15['UC since 2013'].str.replace('U2', '')
df15 = pd.DataFrame(data=df15)

#16. PU - Publisher
df16 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('PU')]
df16.index = np.arange(1,len(df16)+1)
df16.columns = ['Publisher']
df16 = df16['Publisher'].str.replace('PU', '')
df16 = pd.DataFrame(data=df16)


#17. PI - Publisher City
df17 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('PI')]
df17.index = np.arange(1,len(df17)+1)
df17.columns = ['Publisher City']
df17 = df17['Publisher City'].str.replace('PI', '')
df17 = pd.DataFrame(data=df17)

#18. PD - Publication Date
df18 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('PD')]
df18.index = np.arange(1,len(df18)+1)
df18.columns = ['Publication Date']
df18 = df18['Publication Date'].str.replace('PD', '')
df18 = pd.DataFrame(data=df18)


#19. PY - Year Published
df19 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('PY')]
df19.index = np.arange(1,len(df19)+1)
df19.columns = ['Year Published']
df19 = df19['Year Published'].str.replace('PY', '')
df19 = pd.DataFrame(data=df19)

#20. PG - Page Count
df20 = data[data['FN Clarivate Analytics Web of Science'].str.startswith('PG')]
df20.index = np.arange(1,len(df20)+1)
df20.columns = ['Page Count']
df20 = df20['Page Count'].str.replace('PG', '')
df20 = pd.DataFrame(data=df20)


df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17, df18, df19, df20],axis=1)

df.head(10)
df.to_csv('data.csv', encoding='utf-8', index=False)
```
